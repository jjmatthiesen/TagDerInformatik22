```python
import pandas as pd
from matplotlib import pyplot as plt
import scipy.stats as stats
import numpy as np
from IPython.display import display
import seaborn as sns
```

```python
df = pd.read_csv('data/weatherHistory.csv', parse_dates=['Formatted Date'])
print(df.shape)
print(df.info())
```

```python
print(display(df.describe(include='all')))
```

```python
df = df.drop_duplicates(['Formatted Date'], keep='first')
print(df.shape)
```

```python
df.sort_values(by=['Formatted Date'], inplace=True)
df.reset_index(inplace=True, drop=True)
```

```python
display(df)
```

```python
df = df.drop(columns=['Loud Cover', 'Formatted Date', 'Summary', 'Daily Summary'])
print(df.isnull().sum(axis=0) * 100/len(df))
```

```python
df = df[df['Precip Type'].notna()]
```

```python
df.to_csv("data/weather_cleaned01.csv", index=False)
```
